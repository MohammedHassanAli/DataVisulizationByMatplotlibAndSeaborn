```python
tips = pd.read_csv('tips.csv')
```

```python
party_counts = pd.crosstab(tips['day'] , tips['size'])
```

```python
party_counts = party_counts.loc[: ,2:5 ]
```

```python
party_counts
```

```python
party_pcts = party_counts.div(party_counts.sum(1) , axis = 0 )
```

```python
party_pcts.plot.bar()
```

```python
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
```

```python
tips.head()
```

```python
sns.barplot(x='tip_pct' , y = 'day' , orient = 'h' , data = tips)
```

```python
sns.barplot(x='tip_pct' , y = 'day' , hue = 'time' , orient = 'h' , data = tips)
```

```python
tips['tip_pct'].plot.hist(bins=50,grid=True)
```

```python
tips['tip_pct'].plot.density()
```

```python
#scatter or point plot
macro = pd.read_csv('macrodata.csv')
```

```python
macro.head()
```

```python
data = macro[['cpi' , 'm1' , 'tbilrate', 'unemp']]
```

```python
trans_data = np.log(data).diff().dropna()
```

```python
trans_data[-5:]
```

```python
sns.regplot('m1' , 'unemp' , data = trans_data)
```

```python
#pairs ploting in scatter sns
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.2})
```

```python
sns.factorplot(x='day' , y = 'tip_pct' ,hue = 'time' , data = tips[tips.tip_pct < 1], col = 'smoker' , kind='bar')
```
